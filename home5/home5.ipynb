```python
#1. перечень всех документов

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
        
def print_owner():
    result = input('Введите номер документа:')          
    for name in documents:
        if name['number'] == result:
            print('Владелец документа:',name['name']) 
            return name['name']        
    
    print('Документ не найден в базе') 
    return None

#2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

def get_shelf_by_document(doc):         
    for key, value in directories.items():
        for el in value:
            if el == doc:
                return key        
    return None

def print_shelf_by_document():
    result = input('Введите номер документа:')
    shelf = get_shelf_by_document(result)
    
    if shelf is None:
        print('Документ не найден в базе')
    else:
        print(f'Документ хранится на полке: {shelf}') 

# Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам

def print_full_info():
    for el in documents:
        print(f'№: {el["number"]}, тип: {el["type"]}, владелец: {el["name"]}, полка хранения: {get_shelf_by_document(el["number"])}')
        

# Пункт 4. Пользователь по команде “as” может добавить новую полку

def print_shelves():
    return ','.join(directories.keys())

def add_shelf():
    result = input('Введите номер полки:')
    if result in directories:
        print(f'Такая полка уже существует. Текущий перечень полок: {print_shelves()}') 
    else:  
        directories[result] = []
        print(f'Полка добавлена. Текущий перечень полок: {print_shelves()}')
        
   
# Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)

def remove_shelf():
    result = input('Введите номер полки:')
    if result not in directories:
        print(f'Такой полки не существует. Текущий перечень полок: {print_shelves()}')
    else:
        if len(directories[result]) == 0:
            del(directories[result])
            print(f'Полка удалена. Текущий перечень полок: {print_shelves()}')
        else:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {print_shelves()}')
            
while True:
    cmd = input('Введите команду:')
    if cmd == 'p': 
        print_owner()
    elif cmd == 's':
        print_shelf_by_document()
    elif cmd == 'l':
        print_full_info()
    elif cmd == 'as':
        add_shelf()
    elif cmd == 'ds':
        remove_shelf()
    elif cmd == 'q':
        break
```
